```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1.
x_names = ['Java', 'Python','PHP', 'JS','C#','C++']
plt_title = 'Popularity of languages'
y = [22.2,17.6,8.8,8.0,7.7,6.7]
#x_pos = [i for i,_ in enumerate(x_names)]
x_pos = list(range(len(x_names)))
```

```python
plt.bar(x_pos,y, color = 'blue')
plt.xlabel('Language')
plt.ylabel('Percentage')
plt.title(plt_title)
plt.xticks(x_pos,x_names)
plt.minorticks_on()
plt.grid(which='major', linestyle='-',linewidth = 0.5)
plt.grid(which='minor', linestyle=':',linewidth = 0.5)
```

```python
#2.
x_names = ['Java', 'Python','PHP', 'JS','C#','C++']
plt_title = 'Popularity of languages'
y = [22.2,17.6,8.8,8.0,7.7,6.7]
#x_pos = [i for i,_ in enumerate(x_names)]
x_pos = list(range(len(x_names)))
```

```python
#4.
x_names = ['Java', 'Python','PHP', 'JS','C#','C++']
plt_title = 'Popularity of languages'
y = [22.2,17.6,8.8,8.0,7.7,6.7]
#x_pos = [i for i,_ in enumerate(x_names)]
x_pos = list(range(len(x_names)))

plt.bar(x_pos,y,color = ['red','black','green','blue','yellow','cyan'],zorder = 3)
plt.xticks(x_pos,x_names,rotation = 20)
#plt.xticks(xt,xt,rotation = 20)
plt.grid(which = 'major', color = 'red', linewidth = 0.5,linestyle='-',zorder = 0)
```

```python
#5.
x_names = ['Java', 'Python','PHP', 'JS','C#','C++']
plt_title = 'Popularity of languages'
y = [22.2,17.6,8.8,8.0,7.7,6.7]
#x_pos = [i for i,_ in enumerate(x_names)]
x_pos = list(range(len(x_names)))

fig, ax = plt.subplots()
wid = np.random.rand(6) +1 / 3
rects1 = ax.bar(x_pos,y,edgecolor = 'black',width = wid)
plt.xticks(x_pos,x_names,rotation = 20)
plt.subplots_adjust(bottom = -0.1)
plt.xlabel('languages')
#def autolabel(rects):
#ax.text(0.8,2,'HI')
plt.ylim(0,25)
for rect in rects1:
    height = rect.get_height()
    #print(height,rect.get_x()+rect.get_width()/2.)
    ax.text (rect.get_x()+rect.get_width()/2.,1.05*height,f'{height:.1f}',va = 'bottom',ha = 'center')
    
#autolabel(rects1)
#plt.show()
#plt.xticks(xt,xt,rotation = 20)
#plt.grid(which = 'major', color = 'red', linewidth = 0.5,linestyle='-')
```

```python
#12.
menMeans = (54.74, 42.35, 67.37, 58.24, 30.25)
menStd = (4, 3, 4, 1, 5)

index = np.arange(len(menMeans))
widths = 0.35
fig, ax = plt.subplots(1,1,figsize = (10,10))


rects = ax.bar(index,menMeans,yerr=menStd, width = widths,
       color = 'red',zorder = 3)

plt.grid(which='major',linewidth=0.5,zorder = 0)
for i in rects:
    print(i)

for r in rects:
    height = r.get_height()
    ax.text(r.get_x(),
           height *0.9,
           f'{height}',
           ha = 'left', va = 'bottom',zorder = 4)
```

```python
#14
menMean =[22,30,35,35,26]
womenMean = [25,32,30,35,29]
menStd = [4,3,4,1,5]
womenStd = [3,5,2,3,3]
labels = [f'Group {i}' for i in range(len(menMean))]
idx = np.arange(len(menMean))
width = 0.35
```

```python
plt.figure(figsize=(10,10))
plt.bar(idx,menMean,width = width, yerr=menStd,color='red',label= 'Men',zorder = 2)
plt.bar(idx,womenMean,width = width, bottom = menMean,yerr=womenStd,color='green',label= 'Women',zorder = 2)
plt.legend()
plt.xticks(idx,labels)
plt.tight_layout()
plt.grid(which='major',linewidth = 0.5, zorder = 0)
```
