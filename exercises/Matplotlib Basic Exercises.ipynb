## 5. two lines & legend

```python
x = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]
plt.plot(x,y1, label = "line 1") # add label for legend
plt.plot(x,y2, label = "line 2")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('i am legend')
plt.legend() # show legend
```

### 6. Linestyle


```python
x = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]
plt.plot(x,y1, label = "line 1",color='blue',linewidth = 3) # add label for legend
plt.plot(x,y2, label = "line 2",color='red',linewidth = 5)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('i am legend')
plt.legend() # show legend
```

### 7. styles

```python
x = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]
plt.plot(x,y1, label = "line 1",color='blue',linewidth = 3, linestyle = 'dotted') # add label for legend
plt.plot(x,y2, label = "line 2",color='red',linewidth = 5,linestyle = (0,(10,5,5,1)))
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('i am legend')
plt.legend() # show legend
```

### 8. Markers

```python
x = [1,4,5,6,7]
y = [2,6,3,6,3]
plt.plot(x,y,
        label = 'line1',
        color = 'red',
        linestyle = 'dashdot',
        marker = 'o',
        markerfacecolor = 'blue',
        markersize = 12,
        fillstyle = 'full')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
#plt.xlim(1,8)
#plt.ylim(1,8)
plt.axis([1,8,1,8])
#plt.xticks([1,3,5,7,9])

plt.title('Markers')
print(plt.axis())
```

## 13. grid

```python
data = pd.read_csv('ex0114.csv',sep=',',header=0,parse_dates=True,index_col=0)
```

```python
data.columns = ['Close']
```

```python
data_idxed = data['Close']
```

```python
fig = plt.figure()
graph = fig.add_subplot(111)
graph.plot(data_idxed, color = 'red', linewidth=1,marker ='o')
graph.set_xticks(data_idxed.index)
plt.xlabel('Date')
plt.ylabel('Closing Value')
plt.title('Closing of GOOG')
plt.grid(linestyle='-')
plt.axis([data_idxed.index[0].toordinal(),data_idxed.index[-1].toordinal(),min(data_idxed),max(data_idxed)+0.5])
#print(plt.axis())
#data_idxed.index[0].toordinal()
#min(data_idxed)
#data_idxed[-1].toordinal()
```
