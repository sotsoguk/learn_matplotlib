```python
#5.
x1 = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]
plt.plot(x1,y1,label = 'line 1', color = 'blue')
plt.plot(x1,y2,label = 'line 2', color = 'green')
plt.legend()
plt.title('Legend')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
```

```python
#6.
x1 = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]
plt.plot(x1,y1,label = 'line 1', color = 'blue',linewidth = 3)
plt.plot(x1,y2,label = 'line 2', color = 'red', linewidth = 5)
plt.legend()
plt.title('Legend')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
```
